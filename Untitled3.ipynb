```python
import kagglehub
import os
import pandas as pd
path = kagglehub.dataset_download("kandij/mall-customers")
csv_path=os.path.join(path,"Mall_Customers.csv")
dt=pd.read_csv(csv_path)
dt.head()
```

```python
x = dt.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import KMeans
errors = []
for i in range(2,11):
  Kmeans=KMeans(n_clusters=i,init='k-means++')
  Kmeans.fit(x)
  errors.append(Kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,11),errors)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('errors')
plt.grid()
plt.show()
```

```python
kmeans =KMeans(n_clusters=5,init='k-means++')
kmeans.fit(x)
```

```python
dt['group'] = kmeans.predict(x)
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dt = pd.read_csv('/content/groceries.csv', header = None)
dt.head()
```

```python
trnx = dt.apply(lambda x: x.dropna().tolist(), axis=1).tolist()
trnx[0:3]
```

```python
te = TransactionEncoder()
te_ary = te.fit(trnx).transform(trnx)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
freq_items = apriori(df, min_support=0.03, use_colnames=True)
freq_items.tail()
```

```python
rules = association_rules(freq_items,metric='confidence',min_threshold=0.03)
rules.head()
```

```python
print(rules[['antecedents','consequents','lift']].sort_values(by='lift',ascending=False).head(10))
```

```python
product = input("Enter a product: ").lower()
prd_rule = rules[rules ['antecedents'].apply(lambda x: product in x)]
if not prd_rule.empty:
  print(f"since you bought {product}, you would also like:")
  for i in prd_rule ['consequents']:
    print(list(i)[0])
else:
  print("No recommendation found")
```
